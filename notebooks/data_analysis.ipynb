# Data Analysis

This notebook analyzes the datasets on which we want to perform symbolic regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

plt.rcParams['agg.path.chunksize'] = 10000
```

## Read in the data

```python
dataset1 = pd.read_csv('../data/dataset1.csv')
dataset2 = pd.read_csv('../data/dataset2.csv')
dataset3 = pd.read_csv('../data/dataset3.csv')
```

```python
dataset2['log(f(x))'] = np.log(dataset2['f(x)'])
```

## Plot the datasets
### Dataset 1

```python
plt.figure(figsize=(11,7), dpi=200)
plt.scatter(dataset1['x'], dataset1['f(x)'], s=0.2, c='#C71585')
plt.title('Data Points in Dataset 1')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

### Dataset 2

In the below plot, $x_3$ is represented by color.

```python
fig = plt.figure(figsize=(11,7), dpi=200)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(dataset2['x1'], dataset2['x2'], dataset2['log(f(x))'], s=0.3, c=dataset2['x3'], cmap='magma')
ax.set_title('Data Points in Dataset 2')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f(x_1, x_2, x_3)$')
plt.show()
```

### Dataset 3

```python
plt.figure(figsize=(11,7), dpi=200)
plt.scatter(dataset3['x'], dataset3['f(x)'], s=0.2, c='#C71585')
plt.title('Data Points in Dataset 3')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```
